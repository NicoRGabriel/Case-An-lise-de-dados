### **Função para determinar a Idade e poder tirar a média**

```python
import datetime
```

```python
def calcular_idade(data_nascimento):
    hoje = datetime.date.today()
    idade = hoje.year - data_nascimento.year
    if hoje.month < data_nascimento.month or (hoje.month == data_nascimento.month and hoje.day < data_nascimento.day):
        idade -= 1
    return idade
```

```python
df["Nascimento"] = df["Nascimento"].apply(calcular_idade)
```

```python
df = df.rename(columns={"Nascimento":"Idade"})
```

```python
df.fillna({"Idade":df.Idade.mean()}).tail()
```
